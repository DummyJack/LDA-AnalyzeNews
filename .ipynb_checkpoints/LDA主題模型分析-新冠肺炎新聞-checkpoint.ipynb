```python
# 匯入模組
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

from gensim import corpora, models
from gensim.models import CoherenceModel
from preprocess import preprocess
import pyLDAvis.gensim_models
```

```python
raw_data_dir = os.path.join(os.getcwd(), 'data') # 存儲原始資料的目錄
print(raw_data_dir + '\r\n')
```

# 讀檔

```python
filename_temp = pd.Series(['分詞效果'])
file_seg_effect = ''
```

```python
# 檢查data folder下是否有分詞效果excel表

for item_filename in os.listdir(raw_data_dir):
    if filename_temp[0] in item_filename:
        file_seg_effect = item_filename
        continue
if file_seg_effect:
    print("分詞效果excel表:","data\\", file_seg_effect)
else:
    print("分詞效果excel表不存在")
```

```python
# 讀取分詞效果excel表
df = pd.read_excel(os.path.join(raw_data_dir, file_seg_effect))
df = df.drop(['發佈時間', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', '序號'], axis=1)
```

```python
df.sample(10)
```

```python
# 分詞效果excel表裡的分詞資料用split()就能分成詞數組
df["分詞資料"] = df['分詞資料'].str.split(' ').tolist()

# 把每一行有NaN取代為空字串
df["分詞資料"] = df['分詞資料'].replace(np.nan, '')

df["分詞資料"]
```

# 預處理

```python
# 分詞資料(語料庫)
texts = df['分詞資料'].tolist()

# 去除純數字、一个字的詞
processed_corpus = [[token for token in text if not (token.isnumeric() or len(token) < 2)] for text in texts]

# 顯示前10行的處理结果
for item in range(10):
    print(processed_corpus[item])
```

# LDA主題模型

```python
dic = corpora.Dictionary(processed_corpus) # 根據處理後的語料庫生成唯一性詞典

corpus = [dic.doc2bow(text) for text in processed_corpus] # 文檔詞頻矩陣
```

```python
# TF-IDF
tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```python
Lda_Model = models.LdaModel # 創建Lda對象 
```

# 選擇最佳主題數

## 困惑度(Perplexity)

```python
def lda_model_perplexity(num_topics):
    lda_model = Lda_Model(corpus, num_topics=num_topics, id2word=dic)
    print(lda_model.print_topics(num_topics=num_topics, num_words=10)) # num_words:每個主題下的詞語個數
    print(lda_model.log_perplexity(corpus))   
    return lda_model.log_perplexity(corpus)

def lda_model_tfidf_perplexity(num_topics):
    lda_model_tfidf = Lda_Model(corpus_tfidf, num_topics=num_topics, id2word=dic)
    print(lda_model_tfidf.print_topics(num_topics=num_topics, num_words=10)) # num_words:每個主題下的詞語個數
    print(lda_model_tfidf.log_perplexity(corpus))   
    return lda_model_tfidf.log_perplexity(corpus)
```

```python
x = range(1,51)  # 主題範圍:1-50
y = [lda_model_perplexity(i) for i in x]
```

```python
plt.plot(x, y)
plt.xlabel('Number of Topic')
plt.ylabel('Perplexity')
plt.title('Topic 1~50 / lda_model_Perplexity')
plt.savefig('儲存LDA主題模型圖表\\困惑度和連貫性\\lda_model_Perplexity.png') #儲存圖片
plt.show()
```

```python
x1 = range(1,51)  # 主題範圍:1-50
y1 = [lda_model_tfidf_perplexity(i) for i in x1]
```

```python
plt.plot(x1, y1)
plt.xlabel('Number of Topic')
plt.ylabel('Perplexity')
plt.title('Topic 1~50 / lda_model_tfidf_Perplexity')
plt.savefig('儲存LDA主題模型圖表\\困惑度和連貫性\\lda_model_tfidf_Perplexity.png') #儲存圖片
plt.show()
```

## 主題連貫性(Coherence)

```python
def lda_model_coherence(num_topics):
    lda_model = Lda_Model(corpus, num_topics=num_topics, id2word = dic)
    coherence_model_lda = CoherenceModel(model=lda_model, texts=processed_corpus, dictionary=dic, coherence='c_v') # num_words:每個主題下的詞語個數
    coherence_lda_model = coherence_model_lda.get_coherence()
    print('主題連貫性: %f' %(coherence_lda_model))
    return coherence_lda_model

def lda_model_tfidf_coherence(num_topics):
    lda_model_tfidf = Lda_Model(corpus_tfidf, num_topics=num_topics, id2word = dic)
    coherence_model_lda_tfidf = CoherenceModel(model=lda_model_tfidf, texts=processed_corpus, dictionary=dic, coherence='c_v') # num_words:每個主題下的詞語個數
    coherence_lda_model_tfidf = coherence_model_lda_tfidf.get_coherence()
    print('主題連貫性: %f' %(coherence_lda_model_tfidf))
    return coherence_lda_model_tfidf
```

```python
a = range(1,51)  # 主題範圍:1-50
b = [lda_model_coherence(i) for i in a]
```

```python
plt.plot(a, b)
plt.xlabel('Number of Topic')
plt.ylabel('Coherence')
plt.title('Topic 1~50 / lda_model_Coherence')
plt.savefig('儲存LDA主題模型圖表\\困惑度和連貫性\\lda_model_Coherence.png') #儲存圖片
plt.show()
```

```python
a1 = range(1,51)  # 主題範圍:1-50
b1 = [lda_model_tfidf_coherence(i) for i in a1]
```

```python
plt.plot(a1, b1)
plt.xlabel('Number of Topic')
plt.ylabel('Coherence')
plt.title('Topic 1~50 / lda_model_tfidf_Coherence')
plt.savefig('儲存LDA主題模型圖表\\困惑度和連貫性\\lda_model_tfidf_Coherence.png') #儲存圖片
plt.show()
```

# 模型選擇: 詞袋語料庫訓練的LDA 模型

## 理解模型中每個主題想要表達的意思

```python
# 對原資料進行處理
df["原資料"] = df['原資料'].str.split(' ').tolist()
df["原資料"] = df['原資料'].replace(np.nan, '')
texts = df['原資料'].tolist()
```

```python
def format_topics_sentences(ldamodel=lda_model, corpus=corpus, texts=texts):
    sent_topics_df = pd.DataFrame()

    for i, row in enumerate(ldamodel[corpus]):
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0: 
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords'] # Perc_Contribution:文檔中主題的百分比貢獻

    contents = pd.Series(texts)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)

df_topic_sents_keywords = format_topics_sentences(ldamodel=lda_model, corpus=corpus, texts=texts)
```

```python
# 找到主題最有貢獻的內容

sent_topics_sorted_df = pd.DataFrame()

sent_topics_outdf_grpd = df_topic_sents_keywords.groupby('Dominant_Topic')

for i, grp in sent_topics_outdf_grpd:
    sent_topics_sorted_df = pd.concat([sent_topics_sorted_df, 
                                             grp.sort_values(['Perc_Contribution'], ascending=[0]).head(1)], axis=0)
    
sent_topics_sorted_df.reset_index(drop=True, inplace=True)
sent_topics_sorted_df.columns = ['主題', "主題貢獻度(百分比)", "關鍵字", "內容"]

sent_topics_sorted_df.head(10)
```

## 瞭解主題的數量和分佈

```python
topic_counts = df_topic_sents_keywords['Dominant_Topic'].value_counts()

topic_contribution = round(topic_counts/topic_counts.sum(), 4)

topic_num_keywords = df_topic_sents_keywords[['Dominant_Topic', 'Topic_Keywords']]

df_dominant_topics = pd.concat([topic_num_keywords, topic_counts, topic_contribution], axis=1)

df_dominant_topics.columns = ['主題', '主題的關鍵字', '文檔的數量', '佔文檔多少%']

df_dominant_topics.head(10)
```
