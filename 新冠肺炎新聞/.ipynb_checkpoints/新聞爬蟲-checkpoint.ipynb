```python
# 匯入爬蟲套件

import requests
import json
from bs4 import BeautifulSoup

from tomorrow import threads

import pandas as pd
import re
from datetime import datetime

import os
```

```python
# 加入使用者資訊模擬真實瀏覽網頁的情況
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'}
```

```python
# 抓取新聞
def getNews(response):
    soup = BeautifulSoup(response.text)
    url = soup.find('link')['href']
    ndf = pd.DataFrame(data = [{'標題':soup.find('h1', attrs={'class':'article-title'}).text,
                                '時間':datetime.strptime(soup.find('meta', attrs={'property':'article:published_time'})['content'],'%Y-%m-%dT%H:%M:%S+08:00'),
                                '分類':soup.find('meta',attrs={'property':'article:section'})['content'],
                                '描述':soup.find('meta',attrs={'name':'description'})['content'],
                                '內容':'\n'.join(i.text for i in soup.find('div',attrs={'class':'article-body'}).find_all('p')),
                                '關鍵字':soup.find('meta',attrs={'name':'keywords'})['content'],
                                '來源':soup.find('meta',{'name':'publisher'})['content'],
                                '網址':soup.find('meta', {'property':'og:url'})['content']}],
                       columns = ['標題', '時間', '分類', '描述', '內容','關鍵字', '來源', '網址']) 
    return ndf
```

```python
# 抓取新聞的連結
def getLinks(response):
    links = []
    soup = BeautifulSoup(response.text)
    for i in soup.find_all('h3'):
        url = i.find('a')['href']
        links.append(url)
    return links
```

```python
# 開啟多線程功能
@threads(5)
def MultiThread_Crawl(url):
    try:
        return requests.get(url, headers=headers)
    except:
        pass
```

```python
# 爬取多個分頁的新聞連結並得到抓取新聞的標題、時間、分類、描述、內容、關鍵字、來源、網址
def crawlNews(keywords, pages):
    links = []
    for i in range(pages):
        url = 'https://www.chinatimes.com/search/{}?page={}'.format(keywords, i+1)
        resp = requests.get(url)
        links += getLinks(resp)
        links = list(set(links))  
        print('{} links in page {}.'.format(len(links),str(i)))

    responses = [MultiThread_Crawl(link) for link in links]

    list_of_dataframes = []
    for response in responses:
        try:
            ndf = getNews(response)
            list_of_dataframes.append(ndf)
        except:
            pass
    df = pd.concat(list_of_dataframes, ignore_index=True)
    print('{} News in DataFrame.'.format(len(df)))
    return df
```

```python
%%time
df = crawlNews(keywords='新冠肺炎', pages=1920)
```

```python
df = df.drop(['時間', '分類','來源', '網址'], axis=1)
df
```
